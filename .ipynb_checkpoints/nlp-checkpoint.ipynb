```python
nlp = spacy.load("en_core_web_sm")
text = "West Chestertenfieldville was referenced in Mr. Deeds."
doc = nlp(text)
#by default all is marked as names
for ent in doc.ents:
    print (ent.text, ent.label_)
```

```python
#pipeline overwrite
ruler = nlp.add_pipe("entity_ruler", before="ner")
patterns = [
            {"label": "GPE", "pattern": "West Chestertenfieldville"},
            {"label": "FILM", "pattern": "Mr. Deeds"}
]
ruler.add_patterns(patterns)
nlp.analyze_pipes()
```

```python
doc = nlp(text)
for ent in doc.ents:
    print (ent.text, ent.label_)
```

```python
###### Creating a model from scratch
text = "This is a sample number (555) 555-5555."
nlp = spacy.blank("en")
ruler = nlp.add_pipe("entity_ruler")
patterns = [
            {"label": "PHONE_NUMBER",
             "pattern": [
                 {"ORTH": "("},
                {"SHAPE": "ddd"},
                 {"ORTH": ")"},
            {"SHAPE": "ddd"},
                 {"ORTH": "-", "OP": "?"},
                 {"SHAPE": "dddd"}
                                                  ]}
    
]
ruler.add_patterns(patterns)
doc = nlp(text)
for ent in doc.ents:
    print (ent.text, ent.label_)
```

```python
### match estructures in a text based on the meta
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
with open ("./datasets/texts/wiki_mlk.txt", "r") as f:
    text = f.read()
print (text)
```

```python
#### here we get all the different proper nouns and related words to names
matcher = Matcher(nlp.vocab)
pattern = [{"POS": "PROPN", "OP": "+"}]
matcher.add("PROPER_NOUNS", [pattern], greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])
print (len(matches))
for match in matches[:10]:
    print (match, doc[match[1]:match[2]])
```

```python
import json
with open ("./datasets/texts/alice.json", "r") as f:
    data = json.load(f)
text = data[0][2][0]
text = text.replace("`", "'")
print (text)
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"ORTH": "'"},
           {"IS_ALPHA": True, "OP": "+"},
           {"IS_PUNCT": True, "OP": "*"},
           {"ORTH": "'"}
          ]
matcher.add("PROPER_NOUNS", [pattern], greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])
print (len(matches))
for match in matches[:10]:
    print (match, doc[match[1]:match[2]])
```

By adding custom named entities identified through regular expressions, you can expand the coverage of entity recognition in spaCy to include specific patterns or entities that are not covered by the pre-trained models. This approach can be useful when dealing with domain-specific entities or specific patterns of interest in your text data.
